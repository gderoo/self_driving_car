````python
# We create a series of functions which can be used for pre-processing
import cv2

# Normalization between -1,1
def img_norm(x):
    a = np.zeros(x.shape)
    for i in range(x.shape[0]):
        a[i] = (x[i]*1.0-128.0)/128.0
    return a

# Turn to grey
def img_grey(x):
    a = np.zeros((x.shape[0],x.shape[1],x.shape[2],1))
    for i in range(x.shape[0]):
        a[i,:,:,0] = cv2.cvtColor(x[i], cv2.COLOR_BGR2GRAY)
    return a

# Histogram equalization for contrast
def img_hist_grey(x):
    a = np.zeros((x.shape[0],x.shape[1],x.shape[2],1))
    for i in range(x.shape[0]):
        a[i,:,:,0] = cv2.equalizeHist(cv2.cvtColor(x[i], cv2.COLOR_BGR2GRAY))
    return a
def img_hist_rgb(x):
    a = np.zeros(x.shape)
    for i in range(x.shape[0]):
        img_yuv = cv2.cvtColor(x[i], cv2.COLOR_BGR2YUV)
        img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
        a[i] = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)
    return a
````

````python
import tensorflow as tf
from tensorflow.contrib.layers import flatten

def maxpool2d(x, k=2):
    return tf.nn.max_pool(x,ksize=[1, k, k, 1],strides=[1, k, k, 1],padding='SAME')

def conv2d(x, s, st, mu, sigma, n, pad = 'VALID'):
    conv_W = tf.Variable(tf.truncated_normal(shape = s, mean = mu, stddev = sigma), name=n+"_W")
    conv_b = tf.Variable(tf.zeros(s[3]), name=n+"_b")
    x = tf.nn.conv2d(x, conv_W, strides=[1, st, st, 1], padding=pad) + conv_b
    return tf.nn.relu(x)

def flat_layer(x, s, mu, sigma, n, relu = True):
    flat_W = tf.Variable(tf.truncated_normal(shape = s, mean = mu, stddev = sigma), name=n+"_W")
    flat_b = tf.Variable(tf.zeros(s[1]), name=n+"_b")
    x = tf.add(tf.matmul(x, flat_W), flat_b)
    if relu:
        x = tf.nn.relu(x)
    return x
````

````python
def Net(x, hyper_shape, mu, sigma, dropout = True, keep_prob = 0.5, color = 3, nb_conv = 1):    
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    
    # Layer 1: Convolutional
    conv1 = conv2d(x, (5,5,color,hyper_shape[0]), 1,
                   mu, sigma, "conv1")
    for n in range(1,nb_conv):
        conv1 = conv2d(conv1, (5,5,hyper_shape[0],hyper_shape[0]), 1,
                       mu, sigma, "conv1"+"_"+str(n), pad = "SAME")
    
    # Pooling
    conv1 = maxpool2d(conv1)
    
    # Layer 2: Convolutional
    conv2 = conv2d(conv1, (5,5,hyper_shape[0],hyper_shape[1]), 1,
                   mu, sigma, "conv2")
    for n in range(1,nb_conv):
        conv2 = conv2d(conv2, (5,5,hyper_shape[1],hyper_shape[1]), 1,
                       mu, sigma, "conv2"+"_"+str(n), pad = "SAME")

    # Pooling
    conv2 = maxpool2d(conv2)

    # Flatten
    l3 = tf.reshape(conv2, [-1, 5*5*hyper_shape[1]])
    
    # Layer 3: Fully Connected
    l3 = flat_layer(l3, (5*5*hyper_shape[1],hyper_shape[2]), mu, sigma, "l3")
    if dropout:
        l3 = tf.nn.dropout(l3, keep_prob)
    
    # Layer 4: Fully Connected
    l4 = flat_layer(l3, (hyper_shape[2],hyper_shape[3]), mu, sigma,"l4")
    if dropout:
        l4 = tf.nn.dropout(l4, keep_prob)
    
    # Layer 5: Fully Connected
    logits = flat_layer(l4, (hyper_shape[3],43), mu, sigma, "logit", relu = False) 
    
    return logits
````

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
# Here we put the main input parameters
EPOCHS = 200
BATCH_SIZE = 128
rate = 0.0005
mu = 0
sigma = 0.005
dropout = True
keep_prob = 0.5
hyper_shape = (12,36,400,200)
color = 3 # nb of colours for input
nb_conv = 2 # nb of convolutional layers before pooling 

X_train_norm = img_norm(img_hist_rgb(X_train_augm))
y_train_norm = y_train_augm

X_valid_norm = img_norm(img_hist_rgb(X_valid))
X_test_norm  = img_norm(img_hist_rgb(X_test))

x = tf.placeholder(tf.float32, (None, 32, 32, color))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, n_classes)

logits = Net(x, hyper_shape, mu, sigma, dropout, keep_prob, color, nb_conv)
#logits = Net3(x, mu, sigma, keep_prob)
````

````python
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits, one_hot_y)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)
````

````python
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

````python
from sklearn.utils import shuffle
import time

acc = np.zeros((EPOCHS,2))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train_norm)

    i = 0
    stop = False
    
    print("Training...")    
    while (i<EPOCHS) and (not stop):
        startTime = time.time()
        X_train_norm, y_train_norm = shuffle(X_train_norm, y_train_norm)
        for offset in range(0, n_train, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train_norm[offset:end], y_train_norm[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y})
            
        acc[i,0] = evaluate(X_train_norm, y_train_norm)
        acc[i,1] = evaluate(X_valid_norm, y_valid)
        print("EPOCH {}".format(i+1),
              "\t Train Accuracy = {:.3f}".format(acc[i,0]),
              "\t Validation Accuracy = {:.3f}".format(acc[i,1]),
              "\t %.2fs" % (time.time()-startTime))
        
        # We create a stopping condition to optimize the training time
        if i>5:
            stop = (acc[i,1]>=0.96)
        i +=1
        
    saver.save(sess, './netff')
    print("Model saved")
````
